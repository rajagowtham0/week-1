```python
import pandas as pd

# ================================
# Load dataset (update file name)
# ================================
df = pd.read_csv(r"C:\Users\rajak\Downloads\Disease precaution.csv")   # change if already loaded

# ================================
# Identify disease column
# ================================
# Update 'disease' if your column name is different
disease_column = 'Disease'

# ================================
# Unique disease names & count
# ================================
unique_diseases = df[disease_column].unique()
num_unique_diseases = len(unique_diseases)

print("Number of unique diseases:", num_unique_diseases)
print("Unique disease labels:\n", unique_diseases)

# ================================
# Disease-wise count & percentage
# ================================
disease_summary = (
    df[disease_column]
    .value_counts()
    .reset_index()
)

disease_summary.columns = ['Disease', 'Count']

disease_summary['Percentage (%)'] = (
    disease_summary['Count'] / disease_summary['Count'].sum()
) * 100

# ================================
# Display results
# ================================
print("\nDisease-wise distribution:")
display(disease_summary)
```
