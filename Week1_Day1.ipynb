```python
# finding the unique classes or diseases 
import pandas as pd
df = pd.read_csv(r"C:\Users\rajak\Downloads\AI Internship\dermatology_database_1.csv")   
class_column = 'class'
unique_class = df[class_column].unique()
num_unique_class = len(unique_class)
print("Number of unique classes", num_unique_class)
print("Unique class labels:\n", unique_class)
disease_summary = (
    df[class_column]
    .value_counts()
    .reset_index()
)
disease_summary.columns = ['class', 'Count']
disease_summary['Percentage (%)'] = (
    disease_summary['Count'] / disease_summary['Count'].sum()
) * 100
# Display results
print("\nDisease-wise distribution:")
display(disease_summary)
```
