```python
# finding the unique classes or diseases 
import pandas as pd
df = pd.read_csv(r"C:\Users\rajak\Downloads\AI Internship\dermatology_database_1.csv")   
class_column = 'class'
unique_class = df[class_column].unique()
num_unique_class = len(unique_class)
print("Number of unique classes", num_unique_class)
print("Unique class labels:\n", unique_class)
disease_summary = (
    df[class_column]
    .value_counts()
    .reset_index()
)
disease_summary.columns = ['class', 'Count']
disease_summary['Percentage (%)'] = (
    disease_summary['Count'] / disease_summary['Count'].sum()
) * 100
# Display results
print("\nDisease-wise distribution:")
display(disease_summary)
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
classwise_mean = numeric_df.groupby(df['class']).mean()
print(classwise_mean)
```

```python
# helps to group patient with the similar disease while we compute the group wise mean we can be able to estimate the clinical expression of the data
# with this alone grouping cant be much efficient because for the same disease symptoms may vary across the subjects.
```

```python
# finding the age versus disease relationship
import pandas as pd
import numpy as np
# Replace common invalid age entries
df['age'] = df['age'].replace(['?', 'NA', ''], np.nan)
# Convert age to numeric
df['age'] = pd.to_numeric(df['age'], errors='coerce')
# Verify
print("Age dtype after cleaning:", df['age'].dtype)


# Define clinical age groups
age_bins = [0, 12, 19, 35, 50, 65, 100]
age_labels = [
    'Child (0–12)',
    'Adolescent (13–19)',
    'Young Adult (20–35)',
    'Adult (36–50)',
    'Middle-aged (51–65)',
    'Elderly (>65)'
]

# Create age group column
df['age_group'] = pd.cut(
    df['age'],
    bins=age_bins,
    labels=age_labels,
    include_lowest=True
)
# Age-wise disease count
age_class_count = (
    df.groupby(['class', 'age_group'])
      .size()
      .unstack(fill_value=0)
)

print("\nAge-wise patient count per disease class:")
print(age_class_count)
# Age-wise disease percentage
age_class_percentage = (
    age_class_count.div(age_class_count.sum(axis=1), axis=0) * 100
).round(2)
print("\nAge-wise percentage distribution per disease class:")
print(age_class_percentage)
# Most affected age group per disease
most_affected_age_group = age_class_count.idxmax(axis=1)
print("\nMost affected age group for each disease class:")
print(most_affected_age_group)
```

```python
# this age wise disease grouping would help clinicians to understand which age group of people is affected by the particular disease
# more often and also able to ease their diagnostics.
```
