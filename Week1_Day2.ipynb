```python
# finding prominent symptoms per class
import pandas as pd
# excluding irrelevant columns
exclude_cols = ['age', 'class']
symptom_cols = [col for col in df.columns if col not in exclude_cols]
# Compute class-wise mean
class_symptom_mean = (
    df.groupby('class')[symptom_cols]
      .mean()
)
# Extract TOP 3 symptoms per class
top_3_symptoms_per_class = {}

for disease_class in class_symptom_mean.index:
    top_3 = (
        class_symptom_mean.loc[disease_class]
        .sort_values(ascending=False)
        .head(3)
    )
    top_3_symptoms_per_class[disease_class] = top_3
# displaying the results
for disease_class, symptoms in top_3_symptoms_per_class.items():
    print(f"\nDisease Class {disease_class}: Top 3 prominent symptoms")
    for symptom, value in symptoms.items():
        print(f"  {symptom} → Mean severity: {value:.2f}")
```

```python
# class-wise symptom profiling helps to reveal the symptoms that exhibits highest mean severity scores per class
# this indicates their dominant diagnostic relevance for their respective classes
# can easily find the most prominent symptoms per class
```

```python
# create synthetic patient ID
df.insert(
    0,  # insert as first column
    'patient_id',
    ['P{:04d}'.format(i) for i in range(1, len(df) + 1)]
)

# verify
df.head()
```

```python
# patient ID creation helps to provide a unique ID for the instance 
# helps in easier identification and grouping of the values that exist in the rows.
```

```python
# clustering the patient ID based on the eucledian distance
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
# dropping the irrelevant feature columns
exclude_cols = ['patient_id', 'age', 'class']
feature_cols = [c for c in df.columns if c not in exclude_cols]
X = df[feature_cols]
# feature standardisation using standard scalar
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# K-means-based clustering 
kmeans = KMeans(
    n_clusters=6,
    random_state=42,
    n_init=10
)
cluster_labels = kmeans.fit_predict(X_scaled)
#naming the labels
cluster_map = {0:'A', 1:'B', 2:'C', 3:'D', 4:'E', 5:'F'}
df['cluster'] = [cluster_map[i] for i in cluster_labels]
# value count per each cluster
cluster_counts = df.groupby('cluster')['patient_id'].count()
print("Patient count per Kmeans cluster:")
print(cluster_counts)
```

```python
# based on the similarity among the patient ID clusters have been created using the K means
```

```python
# Cluster vs Class
cluster_class = (
    df.groupby(['cluster','class'])
      .size()
      .unstack(fill_value=0)
)
print(cluster_class)
# Percentage distribution
cluster_class_pct = (
    cluster_class.div(cluster_class.sum(axis=1), axis=0) * 100
).round(2)
print("\nCluster-wise class percentage:")
print(cluster_class_pct)
# Dominant class
dominant_class = cluster_class_pct.idxmax(axis=1)
dominance_strength = cluster_class_pct.max(axis=1)
print("\nDominant class per cluster (clinically valid only if >50%):")
for c in cluster_class_pct.index:
    if dominance_strength[c] > 50:
        print(f"Cluster {c} → Class {dominant_class[c]} ({dominance_strength[c]}%)")
    else:
        print(f"Cluster {c} → Mixed phenotype (no dominant class)")
```

```python
# This shows how patients are distributed to each class versus the created cluster group
# from this, we can easily find which cluster of patients falls into which disease class
# This helps in the earlier diagnosis based on their respective symptoms, and they can find their respective disease based on their symptom
# helps clinicians to easily diagnose the patients based on the existing groups
```

```python
# finding similarity score using cosine angle method
import numpy as np
import pandas as pd
# excluding the columns not required for analysis
exclude_cols = ['patient_id', 'age', 'class']
feature_cols = [c for c in df.columns if c not in exclude_cols]
X = df[feature_cols].values
patient_ids = df['patient_id'].values
# Defining the cosine similarity equation method
def cosine_similarity(v1, v2):
    dot = np.dot(v1, v2)
    norm1 = np.linalg.norm(v1)
    norm2 = np.linalg.norm(v2)
    return dot / (norm1 * norm2)
# Compute similarity matrix
n = X.shape[0]
similarity_matrix = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        similarity_matrix[i, j] = cosine_similarity(X[i], X[j])
# Convert to a DataFrame for readability
similarity_df = pd.DataFrame(
    similarity_matrix,
    index=patient_ids,
    columns=patient_ids
)
# Display result
print("Cosine Similarity Matrix (entire dataset):")
print(similarity_df)
```

```python
# This similarity score shows how much the two subjects are
# It reveals the disease class of the subject based on the shared symptoms among the existing records
# people who have a higher similarity score have a higher chance to have same disease
# helps in providing an easier diagnosis based on verifying with existing.
```

```python
# calculating severity score for each subject
import pandas as pd
# feature selecting
exclude_cols = ['patient_id', 'age', 'class']
feature_cols = [c for c in df.columns if c not in exclude_cols]
# calculating the severity score
df['severity_score'] = df[feature_cols].sum(axis=1)
# sort the subjects by severity
df_sorted = df.sort_values(by='severity_score', ascending=False)
print("Top subjects by overall severity:")
print(df_sorted[['patient_id', 'class', 'severity_score']].head(10))
# finding the highest severity for each class
highest_severity_per_class = (
    df_sorted
    .groupby('class')
    .first()
    .reset_index()
)
print("\nHighest severity subject in each class:")
print(highest_severity_per_class[['class', 'patient_id', 'severity_score']])
```
