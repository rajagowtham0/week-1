# Grouping by class

```python
# Grouping the patients per class
import pandas as pd
patients_by_class = (
    df.groupby('class')['patient_id']
      .apply(list)
)
print("Patients grouped by disease class:")
for cls, patients in patients_by_class.items():
    print(f"\nClass {cls}:")
    print(patients)
class_counts = df['class'].value_counts().sort_index()
print("\nNumber of patients in each class:")
print(class_counts)
```

Grouped the patients based on their classes. Helps clinicians to group the patients based on their prominent disease class. This allows the disease-wise comparison and analysis of the patients.

# Finding prominent symptoms for each class

```python
import pandas as pd
# naming the classes based on the clinical data
class_name_map = {
    1: 'Psoriasis',
    2: 'Seborrheic Dermatitis',
    3: 'Lichen Planus',
    4: 'Pityriasis Rosea',
    5: 'Chronic Dermatitis',
    6: 'Pityriasis Rubra Pilaris'
}
df['disease_name'] = df['class'].map(class_name_map)
# column selecting 
exclude_cols = ['patient_id', 'age', 'class', 'disease_name', 'severity_score']
symptom_cols = [c for c in df.columns if c not in exclude_cols]
# calculating mean severity for symptoms
class_symptom_mean = (
    df.groupby(['class', 'disease_name'])[symptom_cols]
      .mean()
)
# extracting the 4 prominent symptoms for each class
print("\nTop 4 Prominent Symptoms for Each Disease:\n")

for (cls, disease), row in class_symptom_mean.iterrows():
    top_4_symptoms = (
        row.sort_values(ascending=False)
           .head(4)
    )
    
    print(f"Class {cls} – {disease}")
    for symptom, value in top_4_symptoms.items():
        print(f"  {symptom}  → Mean severity: {value:.2f}")
    print("-" * 50)
```

Helps to find the prominent symptoms for each of the classes.
Based on the prominent symptoms, the clinicians and patients can know their disease class.
The observed symptoms are generally used to understand the common pattern seen across patients for the particular disease.

# Grouping patients and symptoms based on severity level

```python
import pandas as pd
import numpy as np
# selecting the columns 
exclude_cols = ['patient_id', 'class', 'age']
numeric_feature_cols = (
    df.drop(columns=exclude_cols, errors='ignore')
      .select_dtypes(include=['int64', 'float64'])
      .columns
)
# calculate severity score
df['severity_score'] = df[numeric_feature_cols].sum(axis=1)
# assign severity category within each class
def assign_severity_category(group):
    low_cutoff = group['severity_score'].quantile(0.33)
    high_cutoff = group['severity_score'].quantile(0.66)
    def label(score):
        if score <= low_cutoff:
            return 'Low'
        elif score <= high_cutoff:
            return 'Moderate'
        else:
            return 'High'
    group['severity_category'] = group['severity_score'].apply(label)
    return group
# apply severity categorisation for each class
df = (
    df.groupby('class', group_keys=False)
      .apply(assign_severity_category)
)
# displaying the head 10 values for the sample 
print("Sample severity categorisation:")
print(df[['patient_id', 'class', 'severity_score', 'severity_category']].head(10))
# finding the severity distribution per class
severity_distribution = (
    df.groupby(['class', 'severity_category'])
      .size()
      .unstack(fill_value=0)
)
print("\nSeverity distribution per class:")
print(severity_distribution)
```

This severity level indicates the relative seriousness of the subject within the group.
Grouping the subjects into the severity category helps to prioritise the patients for earlier diagnosis and monitoring.


Finding the sverity of symptoms for each class.

```python
import pandas as pd
exclude_cols = [
    'patient_id', 'age', 'class',
    'severity_score', 'severity_category'
]

symptom_cols = [
    c for c in df.columns
    if c not in exclude_cols and df[c].dtype != 'object'
]
severity_symptom_profile = (
    df.groupby(['class', 'severity_category'])[symptom_cols]
      .mean()
)
TOP_N = 1

print("\nProminent symptoms by severity level per disease:\n")

for (cls, severity_level), row in severity_symptom_profile.iterrows():
    
    top_symptoms = (
        row.sort_values(ascending=False)
           .head(TOP_N)
    )
    
    print(f"Class {cls} | Severity: {severity_level}")
    for symptom, value in top_symptoms.items():
        print(f"  {symptom} → Mean severity: {value:.2f}")
    print("-" * 60)
```

can able to find the prominent disease among the severity group of each class.
