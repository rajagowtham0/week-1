```python
# loading the dataset and printing the top 5 rows
import pandas as pd
df = pd.read_csv(r"C:\Users\rajak\Downloads\AI Internship\dermatology_database_1.csv")
df.head(5)
```

```python
# create synthetic patient ID
df.insert(
    0,  # insert as first column
    'patient_id',
    ['P{:04d}'.format(i) for i in range(1, len(df) + 1)]
)

# verify
df.head()
```

```python
# Grouping the patients per class
import pandas as pd
patients_by_class = (
    df.groupby('class')['patient_id']
      .apply(list)
)
print("Patients grouped by disease class:")
for cls, patients in patients_by_class.items():
    print(f"\nClass {cls}:")
    print(patients)

# ======================================================
# 2. Count patients per class
# ======================================================
class_counts = df['class'].value_counts().sort_index()

print("\nNumber of patients in each class:")
print(class_counts)
```
